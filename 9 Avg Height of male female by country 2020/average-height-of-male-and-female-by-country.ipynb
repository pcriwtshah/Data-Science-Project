```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
file_path = '../input/height-of-male-and-female-by-country-2022/Height of Male and Female by Country 2022.csv'
height_country_data = pd.read_csv(file_path, index_col="Rank")
```

```python
#Country vs height of male barplot
sns.barplot(x = 'Country Name', y = 'Male Height in Ft', data = height_country_data)

#Show the plot
plt.show()
```

```python
male_height5 = height_country_data.nlargest(5, 'Male Height in Ft')
sns.barplot(x = 'Country Name', y = 'Male Height in Ft', data = male_height5)


plt.title('Average male height of top 5 countries')
plt.xticks(rotation = 45)
plt.savefig("Top 5 average male height.jpg")
```

The top country for average male height is Netherlands.

```python
female_height5 = height_country_data.nlargest(5, 'Female Height in Ft')
sns.barplot(x = 'Country Name', y = 'Female Height in Ft', data = female_height5)
plt.title('Average female height of top 5 countries')
plt.xticks(rotation = 45)
plt.savefig("Top 5 average female height.jpg")
```

The top country for average female height is Netherlands.

```python
sns.boxplot(data= height_country_data.loc[:, ["Male Height in Ft", "Female Height in Ft"]])

plt.show()
```

```python
# Histogram 
sns.histplot(height_country_data['Male Height in Cm'], kde=False)

# Add title
plt.title("Histogram of Male Heights")

# Force legend to appear
plt.legend()
```

The distribution of height of a male follows roughly symmetric distribution with mean height around 173.1 cm and standard deviation of 4.95 cm. There don't appear any outlier in this distribution. The height of male varies from 160.13 cm to 183.78 cm.

```python
sns.histplot(height_country_data['Female Height in Cm'], kde=False)
```

The distribution of height of a female follows roughly symmetric distribution with single peak at 161.0 cm and with mean height around 160.94 cm and standard deviation of 4.08 cm. There don't appear any outlier in this distribution. The height of female varies from 150.91 cm to 170.36 cm.

```python

# Histogram 
sns.histplot(height_country_data['Male Height in Cm'], label="Male Height in cm", color="lime", kde=False)
sns.histplot(height_country_data['Female Height in Cm'], label="Female Height in cm", kde=False)

# Setting the figure size
#plt.figure(figsize=(18,10)) 


# Add title and axes label
plt.title("Histogram of Heights of Male and Female")
plt.xlabel('Heights')
plt.ylabel('Frequency')

# Force legend to appear
plt.legend()

plt.show()
```

We can predict a male or a female based on the height. If the height is greater than 170.36 cm, then the individual is a male. If the height is less than 160.13 cm then the individual is a female. If the height is between 160.13 cm and 170.36 cm, then further analysis is necessary.

```python
# Histogram 
sns.histplot(height_country_data['Male Height in Ft'], label="Male Height in Ft", kde=False)
sns.histplot(height_country_data['Female Height in Ft'], label="Female Height in Ft", kde=False)

# Add title and axes label
plt.title("Histogram of Heights of Male and Female")
plt.xlabel('Heights')
plt.ylabel('Frequency')

# Force legend to appear
plt.legend()

plt.show()
```

We can predict a male or a female based on the average height. If the average height is greater than 5.59 ft, then the individual is a male. If the average height is less than 5.25 ft, then the individual is a female. If the height is between 160.13 cm and 170.36 cm, then further analysis is necessary.
